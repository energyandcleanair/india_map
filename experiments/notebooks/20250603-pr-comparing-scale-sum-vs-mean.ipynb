```python
import pandas as pd
```

```python
scale_mean_10k = pd.read_parquet('../../tmp/scale_10k_mean.parquet')
scale_mean_500 = pd.read_parquet('../../tmp/scale_500_mean.parquet')
scale_mean_nominal = pd.read_parquet('../../tmp/scale_nominal_mean.parquet')
```

```python
scale_mean_10k = scale_mean_10k.sort_values(by=["grid_id"])
scale_mean_500 = scale_mean_500.sort_values(by=["grid_id"])
scale_mean_nominal = scale_mean_nominal.sort_values(by=["grid_id"])

assert scale_mean_10k.shape == scale_mean_500.shape
assert scale_mean_10k.shape == scale_mean_nominal.shape
```

```python
diff_10k_to_500 = scale_mean_10k - scale_mean_500
diff_500_to_nominal = scale_mean_500 - scale_mean_nominal
```

```python
diff_10k_to_500
```

```python
diff_10k_to_500.min().min(), diff_10k_to_500.max().max(), diff_10k_to_500.mean().mean()
```
