```python
import numpy as np
import rasterio

block_size = 5
grid_size = 20
datasets = []

value = 0

for _ in range(4):
    arr = np.zeros((grid_size, grid_size), dtype=np.uint8)
    for row in range(0, grid_size, block_size):
        for col in range(0, grid_size, block_size):
            arr[row:row+block_size, col:col+block_size] = value
            value += 1
    datasets.append(arr)
```

```python
datasets
```

```python
dataset_2022_12_30 = datasets[0]
dataset_2022_12_31 = datasets[1]
dataset_2023_01_01 = datasets[2]
dataset_2023_01_02 = datasets[3]

```

```python
dataset_2023_01_02
```

```python
min_value = min(dataset_2022_12_30.min(), dataset_2022_12_31.min(), dataset_2023_01_01.min(), dataset_2023_01_02.min())
max_value = max(dataset_2022_12_30.max(), dataset_2022_12_31.max(), dataset_2023_01_01.max(), dataset_2023_01_02.max())
print(f"Min value: {min_value}, Max value: {max_value}")
```

```python
with rasterio.open("../../pm25ml/collectors/gee/feature_planner__it_assets/dummy_image/2022-12-30.tif", "r+") as src:
    src.write(dataset_2022_12_30, 11)
    src.update_tags(11, STATISTICS_MINIMUM=min_value, STATISTICS_MAXIMUM=max_value)

with rasterio.open("../../pm25ml/collectors/gee/feature_planner__it_assets/dummy_image/2022-12-31.tif", "r+") as src:
    src.write(dataset_2022_12_31, 11)
    src.update_tags(11, STATISTICS_MINIMUM=min_value, STATISTICS_MAXIMUM=max_value)

with rasterio.open("../../pm25ml/collectors/gee/feature_planner__it_assets/dummy_image/2023-01-01.tif", "r+") as src:
    src.write(dataset_2023_01_01, 11)
    src.update_tags(11, STATISTICS_MINIMUM=min_value, STATISTICS_MAXIMUM=max_value)

with rasterio.open("../../pm25ml/collectors/gee/feature_planner__it_assets/dummy_image/2023-01-02.tif", "r+") as src:
    src.write(dataset_2023_01_02, 11)
    src.update_tags(11, STATISTICS_MINIMUM=min_value, STATISTICS_MAXIMUM=max_value)
```
